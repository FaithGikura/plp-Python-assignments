```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
print(data.columns)
```

```python
data.head
```

```python
print("\nData types of each column:")
print(data.dtypes)
```

```python
print("\nChecking for missing values:")
print(data.isnull().sum())
```

```python
diabetes = data.dropna()
```

```python
print("\nDescriptive statistics:")
print(data.describe())
```

```python
grouped_means = diabetes.groupby('Outcome').mean()
print("\nAverage of features for each outcome (0 = Non-Diabetic, 1 = Diabetic):")
print(grouped_means)
```

```python
print("\nObservations:")
print("- Diabetic patients tend to have higher glucose levels.")
print("- Insulin, BMI, and age are typically higher in diabetic patients.")
print("- Non-diabetic patients generally have lower glucose and BMI values.")
```

```python
plt.figure(figsize=(10,6))
plt.plot(diabetes.index, diabetes['Glucose'], label='Glucose Level', color='teal')
plt.title('Glucose Levels Across Samples')
plt.xlabel('Sample Index')
plt.ylabel('Glucose Level')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Outcome', y='BMI', data=diabetes, )
plt.title('Average BMI by Diabetes Outcome')
plt.xlabel('Outcome (0 = Non-Diabetic, 1 = Diabetic)')
plt.ylabel('Average BMI')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(diabetes['Age'], bins=20, color='salmon', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=diabetes, palette='muted')
plt.title('Glucose vs BMI (Colored by Outcome)')
plt.xlabel('Glucose Level')
plt.ylabel('BMI')
plt.legend(title='Outcome')
plt.show()
```
